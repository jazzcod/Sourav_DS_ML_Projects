```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
ds_stu=pd.read_csv("Student_marks.csv")
ds_stu.columns=["Name","Gender","DOB","Maths","Physics","Chemistry","English","Biology","Economics","History","Civics"]
ds_stu
```

```python
ds_stu.info()
```

```python
#checking students who got more than 75% in Maths
ds_stu.loc[ds_stu["Maths"]>=75,["Name","Gender","DOB","Maths"]]
```

```python
#creating a new column for aggregate marks
total_marks=pd.Series()
total_marks=0
subjects=["Maths","Physics","Chemistry","English","Biology","Economics","History","Civics"]
for subject in subjects:
    total_marks=total_marks+ds_stu[subject]
```

```python
ds_stu["Total Marks"]=total_marks
```

```python
ds_stu
```

```python
#Graphical representation of Aggregate marks of Male vs Female
sns.barplot(x="Gender",y="Total Marks",data=ds_stu)
plt.show()
```

```python
sns.barplot(x="Name",y="Total Marks",data=ds_stu)
plt.show()
```
