### Removing Outliers

```python
from scipy.stats import zscore
z_score=abs(zscore(ds_wine))
print(ds_wine.shape)
ds_wine_final=ds_wine.loc[(z_score<3).all(axis=1)]
print(ds_wine_final.shape)
```

### Log Transformation

```python
for col in ds_wine_final:                   
    if ds_wine_final.skew().loc[col]>.55:
        ds_wine_final[col]=np.log1p(ds_wine_final[col])
```

### Seperating Target and Input Variables

```python
x=ds_wine_final.iloc[:,1:]
x
```

```python
y=ds_wine_final.iloc[:,0]
y
```

### Scaling the Input Variables

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
x
```

### Splitting the Input and the Output Variable

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=45,stratify=y)
```

### Importing Metrics

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score
```

### Binarizing the Target Variables and Defining Multiclass function

```python
from sklearn.preprocessing import LabelBinarizer
def multiclass_roc_auc_score(y_test,pred, average="macro"):
    lb = LabelBinarizer()
    lb.fit(y_test)
    y_test = lb.transform(y_test)
    pred = lb.transform(pred)
    return roc_auc_score(y_test, pred, average=average)
```

### Machine Learning

#### K Nearest Neighbors Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(x_train,y_train)
pred=knn.predict(x_test)
print("Accuracy score is:",accuracy_score(y_test,pred))
print("\n")
print("Cross_val_score is:",cross_val_score(knn,x,y,cv=10).mean())
print("\n")
print("Multiclass_roc_auc_score is:",multiclass_roc_auc_score(y_test,pred))
print("\n")
print("Classification_report")
print(classification_report(y_test,pred))
print("\n")
print("Confusion_matirx")
cm=confusion_matrix(y_test,pred)
plt.figure(figsize=(12,8))
sns.heatmap(cm,annot=True)
print(cm)
```

#### SVC

```python
from sklearn.svm import SVC
svc=SVC()
svc.fit(x_train,y_train)
pred=svc.predict(x_test)
print("Accuracy score is:",accuracy_score(y_test,pred))
print("\n")
print("Cross_val_score is:",cross_val_score(svc,x,y,cv=10).mean())
print("\n")
print("Multiclass_roc_auc_score is:",multiclass_roc_auc_score(y_test,pred))
print("\n")
print("Classification_report")
print(classification_report(y_test,pred))
print("\n")
print("Confusion_matirx")
cm=confusion_matrix(y_test,pred)
plt.figure(figsize=(12,8))
sns.heatmap(cm,annot=True)
print(cm)
```

### Appending Models

```python
models = []
models.append(('KNeighborsClassifier',knn))
models.append(('SVC',svc))
models.append(('LogisticRegression',lr))
models.append(('DecisionTreeClassifier',dtc))
models.append(('GaussianNB',gnb))
models.append(('RandomForestClassifier',rfc))
models.append(('GradientBoostingClassifier',gbc))
models.append(('ExtraTreesClassifier',etc))
models.append(('AdaBoostClassifier',abc))
```

### Framing Compare Tables 

```python
Model=[]
ac=[]
cvs=[]
ras=[]
for name, model in models:
    model.fit(x_train, y_train)
    y_pred=model.predict(x_test)
    ac.append(accuracy_score(y_test, y_pred))
    re=cross_val_score(model, x, y, cv=10, scoring='accuracy').mean()
    ras.append(multiclass_roc_auc_score(y_test,y_pred))
    cvs.append(re)
    Model.append(name)
result = pd.DataFrame({'Model': Model, 'Accuracy_score': ac ,'Cross_val_score':cvs,'Roc_auc_score':ras})
result
```
