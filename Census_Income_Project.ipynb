```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
ds_census=pd.read_csv("Census.csv")
ds_census
```

```python
#dropping columns
ds_census.drop("Unnamed: 15",inplace=True,axis=1)
```

```python
ds_census
```

```python
#replacing NaN values
ds_census=ds_census.replace("?",np.nan)
ds_census
```

```python
#label encodibng target variable
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
ds_census["income"]=le.fit_transform(ds_census["income"].astype(str))
ds_census
```

```python
ds_census.isnull().sum()
```

```python
#replacing missing values
for col in ["workclass","occupation","native.country"]:
    ds_census[col].fillna(ds_census[col].mode()[0],inplace=True)
```

```python
ds_census.isnull().sum()
```

```python
#calling unique objects
for column in ds_census.columns:
    if ds_census[column].dtype==object:
        print(str(column)+":"+str(ds_census[column].unique()))
        print(ds_census[column].value_counts())
        print("************************************************")
        print("\n")
```

```python
ds_census.corr()
```

```python
#plotting correlation in a heatmap.
corr_hmap=ds_census.corr()
plt.figure(figsize=(6,6))
sns.heatmap(corr_hmap,annot=True,fmt=".0%",cmap="Blues")
plt.show()
```
