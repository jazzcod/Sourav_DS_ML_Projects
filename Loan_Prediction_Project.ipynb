```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
ds_loan=pd.read_csv("Loan.csv")
ds_loan
```

```python
ds_loan.info()
```

```python
ds_loan.isnull().sum()
```

```python
#using mode to replace categorical data
for col in ["Gender","Self_Employed","Married","Dependents"]:
    ds_loan[col].fillna(ds_loan[col].mode()[0],inplace=True)
```

```python
#using mean to replace numeric data
for col in ["LoanAmount","Loan_Amount_Term","Credit_History"]:
    ds_loan[col].fillna(ds_loan[col].mode()[0],inplace=True)
```

```python
ds_loan.isnull().sum()
```

```python
#calling unique objects
for column in ds_loan.columns:
    if ds_loan[column].dtype==object:
        print(str(column)+":"+str(ds_loan[column].unique()))
        print(ds_loan[column].value_counts())
        print("************************************************")
        print("\n")
```
