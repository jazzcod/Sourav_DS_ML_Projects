### Removing Outliers

```python
#using z-score to remove the outliers from the variables
from scipy.stats import zscore
z_score=abs(zscore(ds_happy))
print(ds_happy.shape)
ds_happy_final=ds_happy.loc[(z_score<3).all(axis=1)]
print(ds_happy_final.shape)
```

### Plotting Distribution

```python
#plotting the distribution of the variables to check skewness
for i in ds_happy_final.columns:
    plt.figure()
    sns.distplot(ds_happy_final[i])
```

### Checking Skewness

```python
#checking skewness of the variables
ds_happy_final.skew()
```

### Log Transformation

```python
#log transformation to make the skewed variables to a normally distributed data.
for col in ds_happy_final:                  
    if ds_happy_final.skew().loc[col]>.55:
        ds_happy_final[col]=np.log1p(ds_happy_final[col])
```

### Replotting Distribution

```python
#replotting the distribution after lof transformation 
for i in ds_happy_final.columns:
    plt.figure()
    sns.distplot(ds_happy_final[i])
```
